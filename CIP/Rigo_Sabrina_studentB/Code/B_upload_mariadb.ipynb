```python
cur.execute("DROP TABLE IF EXISTS B_awardsData_stage3;")
cur.execute("DROP TABLE IF EXISTS B_movieOscars_stage3;")
```

```python
# Create a table
cur.execute("CREATE TABLE B_awardsData_stage3 ("
            "index_Nr INT PRIMARY KEY,"
            "year VARCHAR(255),"
            "oscars_held INT,"
            "category TEXT,"
            "name TEXT,"
            "title VARCHAR(255),"
            "character_name VARCHAR(255),"
            "honorary_statement TEXT,"
            "description TEXT,"
            "note TEXT,"
            "act_dir TEXT,"
            "winner_type TEXT"
            ");")
print("Table is created")
```

```python
cur.execute("CREATE TABLE B_movieOscars_stage3 ("
            "title VARCHAR(255) PRIMARY KEY,"
            "oscars_total INT"
            ");")
print("Second table is created")
```

```python
#load the df awards
awards = pd.read_csv('../Data/B_stage3.csv', low_memory=False, encoding='utf-8')
```

```python
# Replace NaN values with None
awards = awards.astype(object).where(pd.notnull(awards), None)
```

```python
# Looping through the data frame and adding each row to MariaDB
for i, row in awards.iterrows():
    sql = "INSERT INTO B_awardsData_stage3 (index_Nr, year, oscars_held, category, name, title, character_name, " \
          "honorary_statement, description, note, act_dir, winner_type) " \
          "VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
    cur.execute(sql, tuple(row))
    print(f"Record inserted {i}")
conn.commit()
```

```python
#load the awards per movie
awards_movie = pd.read_csv('../Data/B_sum_df.csv', low_memory=False, encoding='utf-8')
```

```python
# Looping through the data frame and adding each row to MariaDB
for i, row in awards_movie.iterrows():
    sql = "INSERT INTO B_movieOscars_stage3 (title, oscars_total) " \
          "VALUES (%s, %s)"
    cur.execute(sql, tuple(row))
    print(f"Record inserted {i}")
conn.commit()
```

```python
cur.execute("ALTER TABLE B_awardsData_stage3 "
            "ADD FOREIGN KEY (title) REFERENCES B_movieOscars_stage3(title);")
print("Foreign key constraint added.")
```

Check if loading was successful:

```python
try:

    cur.execute("SELECT a.year, m.title, m.oscars_total "
                "FROM B_awardsData_stage3 AS a "
                "JOIN B_movieOscars_stage3 AS m "
                "ON a.title = m.title;")

    # Fetch all the rows from the result
    rows = cur.fetchall()

    # Process the retrieved data
    for row in rows:
        year = row[0]
        title = row[1]
        oscars_total = row[2]
    
        # Perform further processing or print the values
        print(f"Year: {year}, Title: {title}, Oscars Total: {oscars_total}")

except mariadb.Error as e:
    print(f"Error: {e}")
```
