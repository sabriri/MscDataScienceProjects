### Impurify data


At least 10 impurities per 100 observations or > 20 impurities for more than 200 observations.

The scraped data already includes some impurities. And since these impurities include mostly a whole column or even more parts of the dataframe, it should be enough. However, there will still be some added impurities so that there are enough different things that need to be cleaned up. 

Already existing impurities: 

- {""} in the Character_name columns
- Sometimes ; in the end of a value
- Names within character_name have []
- There is not only one film_title column but 4. This needs to be separated.

Added impurities: 

- random generated doubles of rows
- change the order for the entries



```python
import pandas as pd
import numpy as np
```

```python
df_B_stage1 = pd.read_csv("../Data/B_stage1.csv", delimiter = ",")
```

```python
df_B_stage1
```

#### 1. Create Duplicates

Create some random duplicates for 2.5% of the dataframe. Save this to the B_stage2 variable.

```python
duplicate_count = int(len(df_B_stage1) * 0.025)
duplicate_count            #would be 64

rows_to_duplicate = df_B_stage1.sample(n=duplicate_count, replace=True)
B_stage2 = pd.concat([df_B_stage1, rows_to_duplicate], ignore_index=True)
B_stage2
```

#### 2. Shuffle the order in the dataframe

```python
B_stage2 = B_stage2.sample(frac=1).reset_index(drop=True)

B_stage2
```

### Export to csv file B_stage2.csv

```python
B_stage2 = B_stage2.to_csv("../Data/B_stage2.csv", index = False)
```
